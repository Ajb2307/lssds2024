## Virtual Environments
[**Keep your Python's isolated from one another**]

<img src="https://blog.debugeverything.com/wp-content/uploads/2021/04/python-virtualenv-project-structure.jpg" alt="drawing" width="400" height="300">

A [virtual environment](https://docs.python.org/3/library/venv.html#venv-def) is a self-contained directory tree that contains a Python installation for a particular version of Python, plus a number of additional packages. Just what we're looking for!

Most Python distributions will include:
* [venv](https://docs.python.org/3/library/venv.html#module-venv) - to create and manage virtual environments
* [pip](https://docs.python.org/3/installing/index.html#installing-index) - to download and manage 3rd party packages (from PyPi by default)

For more information on these tools, you can read the Python tutorial entry on [virtual environments & packages](https://docs.python.org/3/tutorial/venv.html?highlight=pip)

But we can generalize this idea of a Python virtual environment to *all languages*. This is useful because our data science projects may have dependencies outside of Python itself.

#Create a virtual environment with conda:
To create a virtual environment with Conda, write the following command in your terminal:

```sh
conda create --name myenv
```
Here, "myenv" is the name of your new virtual environment. You can replace "myenv" with any name that you prefer for your environment.

**Install a specific version of Python:** If you want to install a specific version of Python in your new environment, you can do so during the creation of the environment. For example, to install Python 3.8 (and other dependences), write the following command in your terminal:
```sh
conda create --name lssds python=3.8 jupyter pandas scikit-learn astropy matplotlib seaborn
```
Activate (or deactivate) the environment: To activate your new environment, write the following command in your terminal:

```sh
conda activate lssds
```
When you're done working in your environment, you can deactivate it by writing the following command:
```sh
conda deactivate
```
Install packages: To install packages in your environment, **you must first activate it** and then you can use conda install. For example, to install numpy, write the following command in your terminal:

```sh
conda install numpy
```
See all your virtual environments: To see all your virtual environments, write the following command in your terminal:

```sh
conda env list
```
Remove a virtual environment: Finally, to remove a virtual environment, write the following command in your terminal:

```sh
conda env remove --name myenv
```
Export the environment to a .yml file: If you want to share your environment with others or want to recreate it later, you can export your environment into a .yml file. Write the following command in your terminal:
```sh
conda env export > environment.yml
```
This will create a new file called "environment.yml" in your current directory, which includes all the information needed to recreate your environment.

Import an environment from a .yml file: If you want to create an environment from a .yml file, you can do so by writing the following command in your terminal:
```sh
conda env create -f environment.yml --name myenv
```

This will create a new environment based on the specifications in the "environment.yml" file.

With these tools, users can:

- Write code from multiple computers and create different versions of their programs.
- Revert changes if they've uploaded problematic code.
- Collaborate with other team members and more.
- They are essential for tracking different versions of code and facilitating collaboration among developers.

## Enabling Collaboration

Git allows you to copy and share repositories with other developers by using the **git push** and **git pull** commands.

The **git push** command lets you share all the commits and files in the current repository with the remote repository:


```
git push [remote repository] [branch name]
```

Use the **git pull** command to copy the content of a remote repository and merge it with your local copy:

```
git pull [remote repository]
```
